```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
mu1=0.15
mu2=0.2
sigma1=0.25
sigma2=0.3
cc=[-1, -0.8, -0.4, 0, 0.4, 0.8, 1]
```

```python
# w=random.random()
# print(w)
def generate_random_weights():
    weights=[]
    for i in range(100):
        weights.append(random.random())
    # print(weights)
    # print(weights[0])
    return weights
```

```python
def muv(w):
    muv=w*mu1+(1-w)*mu2
    return muv
def sigmav(w, cc):
    sigmav=(w**2*sigma1**2+(1-w)**2*sigma2**2+2*w*(1-w)*sigma1*sigma2*cc)**1/2
    return sigmav
```

```python
def calculate_muv_sigmav(weights, cc):
    muv_cc=[]
    sigmav_cc=[]
    for j in range(len(weights)):
        mu=muv(weights[j])
        muv_cc.append(mu)
        sigma=sigmav(weights[j], cc)
        sigmav_cc.append(sigma)
    return muv_cc, sigmav_cc
```

```python
w=generate_random_weights()
```

```python
# cc=-1
c=cc[0]
m, c=calculate_muv_sigmav(w, c)
x_points=np.array(m)
y_points=np.array(c)
plt.plot(x_points, y_points)
```
